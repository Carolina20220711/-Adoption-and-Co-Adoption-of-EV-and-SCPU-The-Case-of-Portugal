```python
def normalize_text(text):
    if pd.isna(text):
        return text
    text = str(text).lower()
    text = ''.join(c for c in unicodedata.normalize('NFD', text) if unicodedata.category(c) != 'Mn')
    text = text.strip()
    text = ' '.join(text.split())
    return text
```

## Fact Tables

```python
def create_quarter_date_key(row):
    year = row['Year']
    quarter = row['Quarter']

    # Map quarter to the first date of the quarter
    if quarter == 1:
        return f"{year}-01-01"  # January 1st = Q1
    elif quarter == 2:
        return f"{year}-04-01"  # April 1st = Q2
    elif quarter == 3:
        return f"{year}-07-01"  # July 1st = Q3
    elif quarter == 4:
        return f"{year}-10-01"  # October 1st = Q4

active_SCPU['QDateKey'] = active_SCPU.apply(create_quarter_date_key, axis=1)
active_SCPU['QDateKey'] = pd.to_datetime(active_SCPU['QDateKey'])
active_SCPU
```

```python
new_SCPU = new_SCPU.copy()

def create_semester_date_key(row):
    year = row['Year']
    semester = row['Semester']

    # Map quarter to the first date of the quarter
    if semester == 1:
        return f"{year}-01-01"  # January 1st = S1
    elif semester == 2 :
        return f"{year}-07-01"  # July 1st = S2

new_SCPU['SDateKey'] = new_SCPU.apply(create_semester_date_key, axis=1)
new_SCPU['SDateKey'] = pd.to_datetime(new_SCPU['SDateKey'])
new_SCPU
```

# Correlation Matrix

```python
sales['Year'] = sales['Date'].dt.year
```

```python
# Sales_EV: BEV or disel/pev or gasoline/phev
sales['Sales_EV'] = np.where(
    sales['Energy'].isin(['BEV', 'Disel/PHEV', 'Gasoline/PHEV']),
    sales['Sales'],
    0
)

# Sales_BEV: BEV
sales['Sales_BEV'] = np.where(
    sales['Energy'] == 'BEV',
    sales['Sales'],
    0
)

#Sales_PHEV: disel/pev or gasoline/phev
sales['Sales_PHEV'] = np.where(
    sales['Energy'].isin(['Disel/PHEV', 'Gasoline/PHEV']),
    sales['Sales'],
    0
)

sales
```

```python
#Check the created new columns
filtered_sales = sales[sales['Energy'].isin(['BEV', 'Disel/PEV', 'Gasoline/PHEV'])]
filtered_sales.head()
```

```python
#Sales
sales_matrix = sales.groupby(['Year', 'District']).agg({
    'Sales': 'sum',
    'Sales_EV': 'sum',
    'Sales_BEV': 'sum',
    'Sales_PHEV': 'sum'
    }).reset_index()

#SCPU
active_SCPU_matrix = active_SCPU_matrix.groupby(['Year', 'District']).agg({
    'SCPU: Nº installations': 'sum',
    'SCPU: total power installed (kW)': 'sum'
    }).reset_index()

#new_SCPU
new_SCPU_matrix = new_SCPU_matrix.groupby(['Year', 'District']).agg({
    'new_SCPU: Nº installations':'sum',
    'new_SCPU: Connection power (kW)': 'sum'
    }).reset_index()
```

```python
merged = pd.merge(sales_matrix, active_SCPU_matrix, how='left', on=['District', 'Year'])
merged = pd.merge(merged, new_SCPU_matrix, how='left', on=['District', 'Year'])
merged.dropna(inplace=True)
merged.reset_index(drop=True, inplace=True)
#merged
```

```python
merged = pd.merge(merged, population, how='left', on=['District'])
merged.dropna(inplace=True)
merged.reset_index(drop=True, inplace=True)
merged
```

```python
merged.drop(columns=['District'], inplace=True)
merged.rename(columns={'SCPU: Nº installations':'active_SCPU: Nº installations', 'SCPU: total power installed (kW)':'active_SCPU: total power installed (kW)'}, inplace = True)
merged
```

```python
corr = merged.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(18, 10))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", vmin=-1, vmax=1, cmap='coolwarm',
            annot_kws={'size': 10},
            cbar_kws={'shrink': 0.8})

plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

plt.title('Correlation Matrix')

plt.savefig(r"C:\Users\20220711\Desktop\Carolina\Final\Images\correlation_matrix.png", dpi=300, bbox_inches='tight')

plt.tight_layout()
plt.show()
```
